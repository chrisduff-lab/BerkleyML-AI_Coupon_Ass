**Context**

We have been engaged to analyse your dataset on coupons delivered to drivers.  There is a various set of coupon offers and characteristics.  Drivers are also differentiated based on serveral attributes, as detailed below.  This analysis provides insight into what coupon on driver conditions lead to the high or low coupon acceptance rate. **Data**

### Data Description
Keep in mind that these values mentioned below are average values.

The attributes of this data set include:
1. User attributes
    -  Gender: male, female
    -  Age: below 21, 21 to 25, 26 to 30, etc.
    -  Marital Status: single, married partner, unmarried partner, or widowed
    -  Number of children: 0, 1, or more than 1
    -  Education: high school, bachelors degree, associates degree, or graduate degree
    -  Occupation: architecture & engineering, business & financial, etc.
    -  Annual income: less than \\$12500, \\$12500 - \\$24999, \\$25000 - \\$37499, etc.
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM
    - Passenger: alone, partner, kid(s), or friend(s)


3. Coupon attributes
    - time before it expires: 2 hours or one day

## Independent Investigation : Coffee House Coupon
The next investigation focused on "Coffee House" coupons, and the characteristics of passengers who accept these.

**Step 1: Create a fresh dataframe with the subset data.**

```python
coffee_coupon_data = data[data["coupon"] == "Coffee House"].copy()  #create fresh copy of data
coffee_coupon_data.to_excel("data/coffee_coupon_data.xlsx")         #output to xls to help investigate data
```

**Step 2: Establish the coupon acceptance rate for the data subset**

```python
coffee_coupon_avg = coffee_coupon_data["Y"].mean().round(2)
print(f"Average acceptance rate for all 'Coffee House' coupons: ", coffee_coupon_avg)
print(f"Count of 'Coffee House' coupon data points:", len(coffee_coupon_data))
```

**Step 3: Start investigating attributes and their relation with coupon acceptance**
Below is an exploration of every attribute, providing a high level indication of its relevance in predicting coupon acceptance.  The following attributes look interesting for further investigation:  
  - _destination:_ 0.36 acceptance for Home, vs 0.58 for No Urgent Place  
  - _passanger:_ 0.44 if travelling Alone, vs 0.6 with Friend(s)  
  - _time:_ 10AM & 2PM were 0.64 & 0.55 vs below avg accepance all other times  
  - _expiration:_ 0.58 for 1d, vs 0.43 for 2hrs  
  - _CoffeeHouse:_ more than 1 looks statistically different from 1 or less  
  - _toCoupon_GEQ25min:_ drops to 0.34 when true  
  - _direction_same:_ an apparent anomoly is that 81% of coffee coupons were offere to people travelling in the opposite  direction.  You'd expect this to be random, and much closer to 50% across a population.  An area of investigation outside our scope. 


```python
# a loop determining the acceptance rate for each catergory in each column
for col in coffee_coupon_data:
    print(f"\n===== {col} =====\n")
    print(
        coffee_coupon_data.groupby(col).agg(
        count = ("Y", "count"),
        accept_rate = ("Y", "mean")).to_string()
    )
```

```python
#generate some binary masks for futher investigation
maskC1 = ((coffee_coupon_data["destination"]=="No Urgent Place") &
          coffee_coupon_data["time"].isin(['10AM','2PM']))
maskC2 = ((coffee_coupon_data["expiration"]=="1d") &
          coffee_coupon_data["CoffeeHouse"].isin(['1~3', '4~8', 'gt8']))
maskC3 = coffee_coupon_data["toCoupon_GEQ25min"].astype(bool)

# create dictionary of descriptions for each mask
mask_desc = {'maskC1' : 'maskC1 : (Destination == No Urgent Place) AND (time == 10am or 2pm)',
             'maskC2' : 'maskC2 : (expiration == 1 day) AND (Coffee House visits more than 1 per month)',
             'maskC3' : 'maskC3 : more than 25min to Coupon = True'
            }

# meld masks into a single df to facilitate plotting
mask_df = pd.DataFrame({
    'maskC1' : maskC1,
    'maskC2' : maskC2,
    'maskC3' : maskC3,
    'Y' : coffee_coupon_data['Y'].astype(bool)  #convert 'Y' to bool for interpretability
})

#define which columns to drill into
keep_columns = ["destination", "passanger", "time", "expiration", "CoffeeHouse", "toCoupon_GEQ25min"]
#define mask list for stepping thorugh 
mask_list = ['maskC1', 'maskC2', 'maskC3']

#step through each mask, plot its relationship with 'Y' and how it relates to other attributes
results_list = []   #initialise the empty result_list[] to capture acceptance rate results 
for mask in mask_list:
    print(f"\n\n\n=====++++++ Investigate {mask} ++++++=====\n")
    print(mask_desc[mask],"\n")
    print(f'Acceptance rate for ', mask, ' : ', 
      coffee_coupon_data[mask_df[mask]]["Y"].mean().round(2))   #print acceptance rate for mask
#    print(f'mask_df checksum: ', mask_df[mask_df[mask]]['Y'].mean().round(2)) # error checking

    #plot the coupons split by the mask, and differentiated by accept and not accept
    plt.figure(figsize=(4, 3))   #set smaller plot size
    sns.countplot(
        data = mask_df,
        x = mask,
        hue = 'Y'
    )
    plt.legend(title="Coupon Accepted")
    plt.title(f'Coupon Acceptance Count for {mask}')
    plt.show()

    # a loop determining the acceptance rate for each catergory in each column
    for col in keep_columns:
        grouped = (
            coffee_coupon_data[mask_df[mask]].groupby(col).agg(
                count = ("Y", "count"),
                accept_rate = ("Y", "mean"))
        ).reset_index() 
        
        #print(f"\n===== {mask} & {col} =====\n")        
        
        grouped["Attribute"] = col + '=' + grouped[col].astype(str)
        grouped["Mask"] = mask
        grouped = grouped.drop(columns=[col])
        grouped = grouped[["Attribute", "Mask", "count", "accept_rate"]]
        
        #print(grouped)
        results_list.append(grouped)
        #print(f'len results_list: ',len(results_list))
    
results = pd.concat(results_list, ignore_index=True)
print(f'\n\n=====++++++ Summary Table for Union of Masks an Attributes ++++++=====\n')
print(results.sort_values("accept_rate", ascending=False))
        
```

## Findings on Coffee Coupons
The average coffee coupon acceptance rate across the population of Coffee Coupons was 0.5.
The count of coffee coupons was 3996.

Every attribute was analysed for its effect on coffee coupon acceptance.  A distilled list of attributes with higher potential to explain coupon exceptance was found to be _passanger, time, expiration, CoffeeHouse, toCoupon_GEQ25min_.  

Further investigation was conducted into the three masks and its unioun with every attribute.  This revealed some very specific sub-groups who could be targeted with a high liklihood of acceptance.

As a general sub-group MaskC2 : (expiration == 1 day) AND (Coffee House visits more than 1 per month) showed an acceptance rate of 0.77.  This aligns with expectations that a longer expiry time provided to people who already visit coffee houses regularly will result in increased uptake.  840 observations fit in this group.

By contrast, MaskC3 : more than 25min to Coupon = True showed a very low acceptance rate of 0.35.  This aligns with an expectation that a driver won't travel 25min or more for a coffee coupon.  

When drilling down, some specific sub-groups revealed which could be targeted.  When maskC2 is combined with any of the attribute conditions in this table, the acceptance rate goes above 0.80.  This suggests a group to target with approppraite coupons.
